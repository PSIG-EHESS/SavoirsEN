```python
# operation system
import os
# regular expression
import re
# retrives file support
import glob
# make a table
import pandas as pd
# regular expression
import re

# import NLP SpaCy
import spacy
nlp = spacy.load("fr_core_news_md")


# inform the path
for file_path in glob.iglob('db_conll/*txt'):
    #transform path into a readable file
    f_name = (file_path)
    #print(f_name)
    
    # open the variable to be read and split into words
    with open(f_name, 'r', encoding='utf8') as f:
        # read and split into words to count word in file
        text = f.read()
        #print(text)
        
        # apply SpaCy
        doc_dc = nlp(text)
        
        # table creator

        # NLP table
        table_view = pd.DataFrame([(e.text, e.ent_iob_, e.ent_type_) for e in doc_dc],  )

        # Enteties column
        table_view["1"] = table_view[1] +'-'+ table_view[2]

        # replace from O- to O 
        table_view['1'].replace( { r"O-" : 'O' }, inplace= True, regex = True)

        # delete the previews columns
        del table_view [1]
        del table_view [2]

        # display entire table
        pd.set_option("display.max_rows", None, "display.max_columns", None, )

        # delete the index
        blankIndex=[''] * len(table_view)
        table_view.index=blankIndex

        # delete the first line
        new_header = table_view.iloc[0] 
        table_view = table_view[1:] 
        table_view.columns = new_header 


        #print(table_view)
        
        # directory out
        output_dir = "SpaCy_md/"
        # new files out with original's name plus _text and its new format .txt
        results_file = "%s%s_SpaCY_md.tsv"%(output_dir, os.path.splitext(os.path.basename(f_name))[0])
        print(results_file)
        # save it as blabla_text.txt
        with open(results_file, 'w') as fpout: 
            fpout.write(str(table_view))
            #print to verify the result
            #print(first)

```
