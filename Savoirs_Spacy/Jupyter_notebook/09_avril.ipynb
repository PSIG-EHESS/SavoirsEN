## Start SpaCy  procedure ##
### import spacy and its model ###

```python
import spacy
nlp = spacy.load("fr_core_news_md")
```

### import the file in its version .txt

```python
with open ("output_savoirs/BNU_02_Colin.txt", "r", encoding="utf_8") as myfile:
    data = myfile.read()
```

### apply SpaCy's model to the file

```python
doc = nlp(data)
```

###  I prefer print it to verify if the file is loaded

```python
print(type(doc))
```

### apply a python function to classify the entities catches from the file

```python
def show_ents(doc): 
    if doc.ents: 
        for ent in doc.ents: 
            print(ent.text+' - ' +str(ent.start_char) +' - '+ str(ent.end_char) +' - '+ent.label_+ ' - '+str(spacy.explain(ent.label_))) 
            print("\n")             
show_ents(doc)
```

### display the results in the file as highlight

```python
from spacy import displacy
displacy.render(doc, style = "ent",jupyter = True)
```

### count the EN on the file

```python
def count_ents(ne_mentions):
    sorted_count_occ = {}
    count_occ = {i:ne_mentions.count(i) for i in ne_mentions}
    #print(count_occ)
    sorted_count_occ = {k: v for k, v in sorted(count_occ.items(), key=lambda item: item[1])}
    #print(sorted_count_occ)
    for k in sorted_count_occ:
        print(k, sorted_count_occ[k])
    return sorted_count_occ
```
