## transformation a file xml into a text file


```python
import re

text = re.sub('<[^<]+>', "", open("output_text/BNU_04_Gfrereis.xml").read())
with open("output_clean/BNU_04_Gfrereis.txt", "w") as f:
    f.write(text)
```
