```python
import spacy
nlp = spacy.load("fr_core_news_md")
```

```python
with open ("output_savoirs/BNU_01_Didier.txt", "r", encoding="utf_8") as myfile:
    data = myfile.read()
```

```python
doc = nlp(data)
```

```python
print(type(doc))
```

```python
def show_ents(doc): 
    if doc.ents: 
        for ent in doc.ents: 
            print(ent.text+' - ' +str(ent.start_char) +' - '+ str(ent.end_char) +' - '+ent.label_+ ' - '+str(spacy.explain(ent.label_))) 
            print("\n")             
show_ents(doc)
```

```python
from spacy import displacy
displacy.render(doc, style = "ent",jupyter = True)
```
